```python
import numpy as np 
import pandas as pd
import re  #text preprocessing lib

# liberay for text preprocessing for stopword removeal,steaming,tokenization
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords # used for stop word removeal . it is a dict of stopwords
from nltk.stem.porter import PorterStemmer #user for steaming the word    

from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.model_selection import train_test_split
import torch
from torch.utils.data import TensorDataset , DataLoader
import torch.nn as nn
import torch.optim as optim
```

# Step-1 Load the Dataset

```python
origional_df=pd.read_csv("IMDB Dataset.csv")
df=origional_df.copy()
```

# Step-2 Data Validation/Checking

```python
#check for null values
df.isnull().sum()
# check for duplicate values
df.drop_duplicates(inplace=True) # so we have dropped nearly 418 duplicate recoeds
df.shape
```

# Step-3 Text PreProcessing 

### 1. convert to lowercasing

```python
df["review"]=df["review"].str.lower()
```

### 2. remove the url's

```python
def removeUrl(text):
    output=re.sub(r"http\S+","",text)
    return output
df["review"]=df["review"].apply(removeUrl)
```

### 3.Remove Html Tags

```python
def removeHtmlTag(text):
    output=re.sub(r"<.*?>","",text)
    return output

df["review"]=df["review"].apply(removeHtmlTag)
```

### 4.Remove Punctuation's eg-> .,#"

```python
def removePunctuation(text):
    output=re.sub(r"[^A-Za-z0-9\s]","",text)
    return output

df["review"]=df["review"].apply(removePunctuation)
```

### 5.Remove Stopwords eg-> a,is,the,and

```python
# from nltk.corpus import stopwords # used for stop word removeal . it is a dict of stopwords
def removeStopWords(text):
    stop_words=stopwords.words("english")
    #firlty we have to convert this text into token's
    tokens=word_tokenize(text)
    # now check if the indivisual token is a stop word or not . if yes them remove it or replace with ""
    for word in tokens:
        if word in stop_words:
            text=text.replace(word,"")

    return text

print("stop word removal started")
df["review"]=df["review"].apply(removeStopWords)
print("stop word removal completed")
```

### 6. Apply Steaming

```python
# from nltk.stem.porter import PorterStemmer #user for steaming the word 
def applySteaming(text):
    tokens=word_tokenize(text)
    steamed_words=[]
    for word in tokens:
        steamed_word=PorterStemmer().stem(word)
        steamed_words.append(steamed_word)
    #return the text Note-> as of now the text in array so convert it into a string and return it 
    return "".join(steamed_words)
print("the steaming has been started")
df["review"]=df["review"].apply(applySteaming)
print("the steaming has been finished")
```

### 7. Encode the Sentiment(output) column

```python
encoder=LabelEncoder()
df["sentiment"]=encoder.fit_transform(df["sentiment"])
Y=df["sentiment"]
```

### 8. Convert the text into vector's using TF-IDF

```python
tfidf=TfidfVectorizer(max_features=2000)
X=tfidf.fit_transform(df["review"])
```

# Text Pre-Processing is completed

# Step-4 Split the data for training and testing

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Step-5 Create TensorDataset & DataLoader

### 1.Convert the data into array from sparses matrics

```python
# convert the sparsed matrized to array so that it can be converted into tensors
x_train=x_train.toarray()
x_test=x_test.toarray()
```

### 2. Create TensorData 

```python
train_set=TensorDataset(torch.from_numpy(x_train).float() ,torch.from_numpy(y_train.values).float())
test_set=TensorDataset(torch.from_numpy(x_test).float(),torch.from_numpy(y_test.values).float())
```

### 3. Create DataLoader from this Tensor Dataset

```python
train_loader=DataLoader(train_set,shuffle=True,batch_size=64)
test_laoder=DataLoader(test_set,batch_size=64)
```

# Step-6 Build the RNN Aircheture

```python
class RNN(nn.Module):
    # create the constructor
    def __init__(self,input_size,hidden_layer=128,num_layer=1):
        # call the super construcot
        super().__init__()
        
        self.input_size=input_size
        self.hidden_layer=hidden_layer
        self.num_layer=num_layer

        # create the RNN layer
        self.rnn=nn.RNN(input_size,hidden_layer,num_layer,batch_first=True)

        # FC layer
        self.fc=nn.Linear(hidden_layer,1)

    def forward(self,x):
        #             no of layers, batch_size , hideen_size
        h0=torch.zeros(self.num_layer,x.size(0),self.hidden_layer)
        # feed the rnn with x and timestamp
        out,_=self.rnn(x,h0)
        final_out=self.fc(out[:,-1,:])
        return final_out
```

# Step-7 Initilize the RNN model

```python
input_size=x_train.shape[1]
model=RNN(input_size)
optimizer=optim.Adam(model.parameters())
createrion=nn.BCELoss()
```

# Step-8 Train the Model

```python
epochs=100

epoc_training_loss=0
for epoc in range(epochs):
    training_loss=0
    model.train()
    for xb,yb in train_loader:
        optimizer.zero_grad()
        outputs=model(xb.unsqueeze(1))  #unsquess as the rnn required a 3d vector
        # before calcuationg loss reduce the incredse dimention
        # this final_out is a logits tensor convert . apply a activaitn fxn to convert it into probabilites
        final_out=torch.sigmoid(outputs.squeeze())
        # print(final_out.shape)
        # print(yb.shape)
        loss=createrion(final_out,yb) # calcualte the loss
        loss.backward()               # compute the gradients
        optimizer.step()              # update the weights 
        training_loss+=loss.item()
    final_loss=training_loss/len(train_loader)
    print(f"epoc no is {epoc+1} the loss is {final_loss} ")
```

# Step-9 Evaluate the model 

```python
model.eval()
total_val=0
correct_val=0
with torch.no_grad():
    for xb,yb in test_laoder:
        output=model(xb.unsqueeze(1))
        final_out=torch.sigmoid(output.squeeze())
        predicted=(final_out>0.5).float()
        total_val+=yb.size(0)
        correct_val+=(predicted==yb).sum().item()

print(f"the correct pred are {correct_val}")
print(f"the total pred are {total_val}")
```
